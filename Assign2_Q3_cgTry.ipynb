```python
#Created for google colab.

!pip install ColabTurtle  #only need to run once. not a big deal to run every time though

import ColabTurtle.Turtle as turtle

###################################
# Set up

turtle.initializeTurtle()
turtle.speed(10)       # fastest for collab
turtle.pencolor("purple")
turtle.pensize(2)

####################################
# Recursive fractual  ----/\----- segmented line
def draw_line(length, depth):
    if depth == 0:    #this is base case
        turtle.forward(length)
    else:
        length = length / 3
        depth = depth - 1
        draw_line(length, depth)
        turtle.right(60)
        draw_line(length, depth)
        turtle.left(120)
        draw_line(length, depth)
        turtle.right(60)
        draw_line(length, depth)

###################################
# Polygon with fractual lines
# this function sets out the number of sides to place the segmented line
def draw_shape(sides, length, depth):
    angle = 360 / sides
    for _ in range(sides):
        draw_line(length, depth)
        turtle.right(angle)

##################################
#User input parameters

sides = int(input("Enter the number of sides: "))
length = int(input("Enter the side length: "))
depth = int(input("Enter the recursion depth: "))


turtle.penup()  # Move to start
#turtle.setpos(-length//2, length//3)    #standard turtle, dosnt work with colab
turtle.setpos(200, 400)  #co-ordinates for google gollab to work. collab cant handle -ve co-ordinates
turtle.pendown()

# Draw
draw_shape(sides, length, depth)
```
